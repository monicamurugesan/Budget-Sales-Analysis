```python
bdo=t2[['CustomerKey','SalesOrderLineNumber','OrderDate']].drop_duplicates()
```

```python
bdo
```

```python
bdo['OrderMonth']=bdo['OrderDate'].dt.to_period('M')
bdo
```

```python
bdo['coho']=bdo.groupby('CustomerKey')['OrderDate']\
                               .transform('min')\
                               .dt.to_period('M')
```

```python
bdo['coho']
```

```python
from operator import attrgetter
bd1=bdo.groupby(['coho','OrderMonth'])\
    .agg(n_customers=('CustomerKey','nunique'))\
    .reset_index(drop=False)
bd1['period_number'] = (bd1.OrderMonth- bd1.coho).apply(attrgetter('n'))
```

```python
bd1
```

```python
cohortpiv=bd1.pivot_table(index='coho',columns='period_number',values='n_customers')
```

```python
cohortpiv
```

```python
cohortsize=cohortpiv.iloc[:,0]
retention_matrix=cohortpiv.divide(cohortsize,axis=0)
```

```python
retention_matrix
```

```python
# plot the retention matrix
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(20, 20), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix, 
                mask=retention_matrix.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='BrBG', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention', fontsize=16)
    ax[1].set(xlabel='# of periods',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohortsize).rename(columns={0: 'cohort_size'})
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                ax=ax[0])

    fig.tight_layout()
```
